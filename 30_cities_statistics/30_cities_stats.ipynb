```python
#overall cities information
cities_stats = baseline_data_accepted.groupby(['region', 'status']).agg(
    total_orders = ('order_id', 'nunique'),
    total_drivers = ('driver_id', 'nunique'),
    avg_wage=('wage', 'mean'), 
    avg_basked=('basket_size', 'mean'), 
    avg_drop_distance=('drop_off_distance', 'mean'), 
    avg_orders_daily = ('total_orders','mean'),
    avg_hours_daily = ('total_hours_worked','mean'), 
).reset_index() 

#acceptance rate per city
acc_rate = baseline_data.groupby('region')['is_driver_accept'].mean().reset_index()
acc_rate = acc_rate.rename(columns={'is_driver_accept':"acceptance_rate"}) 

#all cities data 
cities_info = cities_stats.merge(acc_rate, on='region')

#split into 3 quartiles based on acceptance rate 
p33, p67 = cities_info["acceptance_rate"].quantile([0.33, 0.67])
cities_info["city_type"] = pd.cut(
    cities_info["acceptance_rate"],
    bins=[-np.inf, p33, p67, np.inf],
    labels=["Q1", "Q2", "Q3"],
)

cities_info.head()
```

```python
#helper function to get all the statistical metrics we want 
def describe_full(series, label): 
    s = series.dropna() 
    return pd.DataFrame([{
        'metric': label, 
        'n' : len(s), 
        'mean': s.mean(), 
        'std' : s.std(),
        "min": s.quantile(0.25), 
        'median': s.median(), 
        'Q3': s.quantile(0.75), 
        'max': s.max()
    }])
```

```python
daily = baseline_data_accepted.groupby(['driver_id', 'date', 'Rider Group', 'region', 'status']).agg(
    num_daily_orders=('daily_order_count','max'),
    total_hours_worked=('total_hours_worked', 'max'),
    total_wage=('wage', 'sum')
).reset_index()

days_worked = baseline_data_accepted.groupby('driver_id').agg(
    num_days_worked = ('date', 'nunique')).reset_index()

daily = daily.merge(days_worked, on=['driver_id'], how='left')

daily
```

```python
daily.to_csv('./results/daily_driver_stats.csv')
```

```python
stats = []

for (region, rider_group, status), group in daily.groupby(['region', 'Rider Group', 'status']):
    for metric, label in [
        ('num_days_worked',    'Days worked'),
        ('num_daily_orders',   'Daily orders'),
        ('total_hours_worked', 'Daily hours worked'),
        ('total_wage', 'Daily Wage')
    ]:
        row = describe_full(group[metric], label)
        row['region']     = region
        row['rider_group'] = rider_group
        row['status'] = status
        stats.append(row)

regional_drivers = baseline_data_accepted.groupby('region')['driver_id'].nunique().reset_index()
regional_drivers.columns = ['region', 'total_regional_drivers']

#get group counts of number of drivers per tier per region
group_counts = baseline_data_accepted.groupby(['region', 'Rider Group'])['driver_id'].nunique().unstack(fill_value=0)

group_counts.columns = [f'num_{col.lower()}_drivers' for col in group_counts.columns]

#Reset index so 'region' becomes a column for merging
group_counts = group_counts.reset_index()

stats = pd.concat(stats, ignore_index=True)
stats = stats.merge(regional_drivers, on = 'region', how = 'left')
stats = stats.merge(group_counts, on='region', how='left')

stats
```
