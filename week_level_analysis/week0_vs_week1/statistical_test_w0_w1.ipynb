```python
#test asymmetric response
import statsmodels.formula.api as smf
from scipy import stats
```

```python
week0_week1 = pd.read_csv('./week0_vs_week1_results/driver_summary_wk0_wk1.csv')
```

```python
#outcome variables to test
outcomes = ['orders_completed', 'acceptance_rate', 'hours_worked', 'total_wage']
```

```python
# Stack week0 and week1 at driver level
df = week0_week1.copy()
df['post'] = df['date'].isin(['2025-07-14', '2025-07-15', '2025-07-18', '2025-07-19',
       '2025-07-20', '2025-07-17', '2025-07-16']).astype(int)

# Merge baseline
df = df.merge(week0_baseline, on=['driver_id', 'region', 'Rider Group'], how='left')
df['earnings_gap_to_ref'] = df['total_wage'] - df['ref_wage']

# For week1 rows, classify direction based on incentive vs baseline
df['is_paydown'] = (df['post'] == 1) & (df['earnings_gap_to_ref'] < 0)
df['is_payup']   = (df['post'] == 1) & (df['earnings_gap_to_ref'] > 0)

df['incentive_direction'] = df['earnings_gap_to_ref'].apply(lambda x: 'payup' if x > 0 else 'paydown')
```

```python
for direction, subset in [('paydown', df[df['incentive_direction']=='paydown'].copy()),
                           ('payup',   df[df['incentive_direction']=='payup'].copy())]:
    for outcome in outcomes:
        # Create a temporary subset with no NaNs for the variables in the model
        vars_in_model = [outcome, 'post', 'region', 'avg_rain', 'driver_id', 'Rider Group'] #verify columns exist 
        reg_data = subset[vars_in_model].dropna()

        model = smf.ols(
            f"{outcome} ~ post + C(region) + C(Q('Rider Group')) + avg_rain",
            data=reg_data
        ).fit(cov_type='cluster', cov_kwds={'groups': reg_data['driver_id']})
        
        coef = model.params['post']
        pval = model.pvalues['post']
        print(f"{direction} | {outcome}: coef={coef:.4f}, p={pval:.4f}")
```

```python
df['region_tier'] = df['region'].astype(str) + "_" + df['Rider Group'].astype(str)

for outcome in outcomes:
    formula = f"{outcome} ~ post : C(region_tier) + avg_rain + C(region_tier)"   
    model_df = df.dropna(subset=[outcome, 'post', 'region_tier', 'avg_rain'])
    
    model = smf.ols(formula, data=model_df).fit(
        cov_type='cluster', 
        cov_kwds={'groups': model_df['driver_id']}
    )
    
    print(f"\n--- Results for {outcome} ---")
    cohort_effects = model.params[model.params.index.str.contains('post:')]
    cohort_pvalues = model.pvalues[model.params.index.str.contains('post:')]
    # print(model.summary().tables[1])

    print("Top 5 Group Responses (Post-Incentive Change):")
    print(cohort_effects.sort_values(ascending=False).head(5))
    
```

```python
df_week1 = df[df['date'].isin(['2025-07-14', '2025-07-15', '2025-07-18', '2025-07-19',
       '2025-07-20', '2025-07-17', '2025-07-16'])]
df_week1
```

```python
df_week1['gap'] = df_week1['total_wage'] - df_week1['ref_wage']
df_week1['gap_neg'] = df_week1['gap'].clip(upper=0)   # loss region
df_week1['gap_pos'] = df_week1['gap'].clip(lower=0)   # gain region
```

```python
import numpy as np
```

```python
for outcome in outcomes:
    model_vars = [outcome, 'gap_neg', 'gap_pos', 'region', 'avg_rain', 'driver_id']
    
    subset = df_week1[model_vars].dropna()
    
    # fit the model
    model = smf.ols(
        f"{outcome} ~ gap_neg + gap_pos + C(region) + avg_rain",
        data=subset
    ).fit(cov_type='cluster', cov_kwds={'groups': subset['driver_id']})
    
    # results
    coef_neg = model.params.get('gap_neg', np.nan)
    coef_pos = model.params.get('gap_pos', np.nan)
    
    print(f"\n--- {outcome} ---")
    print(f"  Loss slope: {coef_neg:.4f} | Gain slope: {coef_pos:.4f}")
    
    # tests if the two slopes are statistically different
    wald_test = model.f_test("gap_neg = gap_pos")
    print(f"  Kink test p-value: {wald_test.pvalue:.4f}")
```
