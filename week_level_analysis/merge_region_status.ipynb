```python
import pandas as pd
import os 
```

```python
reference_df = pd.read_csv('../region_status.csv')
```

```python
#get mapping of region -> [Win, Lose, Swing]
mapping = reference_df.set_index('region')['status']
mapping = reference_df.drop_duplicates(subset='region').set_index('region')['status']
mapping
```

```python
folder_path_1 = '../progress_metrics/results'
folder_path_2 = '../progress_metrics/daily_guarantee/daily_guarantee_results'
folder_path_3 = '../mainexperiment/weekly/tier_merged'

output_path_1 = '../progress_metrics/results_updated'
output_path_2 = '../progress_metrics/daily_guarantee/daily_guarantee_results_updated'
output_path_3 = '../mainexperiment/weekly/tier_merged_updated'
```

```python
#add to all incentives datasets
for filename in os.listdir(folder_path_1): 
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join(folder_path_1, filename)) 
        df['status'] = df['region'].map(mapping) 
        # df = df.drop(['Unnamed: 0'], axis =1)
        df.to_csv(os.path.join(output_path_1, filename), index=False)
```
