```python
# Driver-level summary
driver_summary = incentive_progress.groupby(['driver_id', 'Rider Group', 'region', 'date', 'Incentive', 'Status']).agg(
    orders_completed=('daily_order_count', 'max'),  #gets the number of orders completed in a day
    acceptance_rate=('is_driver_accept', 'mean'), #how many orders does the driver accept compared to how many orders they get assigned
    hours_worked=('daily_cumulative_time_hr', 'max'), #how much did do they work a day (week 1)
    total_wage=('wage', 'sum'), #how much money do they make from the order directly (not including on_top_fare; need to clarify what on_top_fare is)
    avg_time_per_order_sec=('time_per_order_sec', 'mean'), #how much time they spend per order on average
    avg_drop_distance=('drop_off_distance', 'mean'), #how far do they go to drop off the order (further or closer locations preferred)
    closest_drop_distance=('drop_off_distance', 'min'), 
    furthest_drop_distance=('drop_off_distance', 'max'), 
    avg_rain = ('avg_rain', 'mean'), #does weather have an impact 
    avg_basket_size=('basket_size', 'mean'),
).reset_index()

driver_summary
```

```python
#plotting config
sns.set_theme(style="whitegrid", font_scale=1.3)

TIER_ORDER = ["Top", "Mid", "Bottom"]
INCENTIVES = ["Guarantee", "No Incentive", "Active Bonus", "Streak", "Daily Productivity"]
STATUS = ['Win', 'Lose', 'Swing']

metrics = {
    "hours_worked":     "Hours Worked / Day",
    "orders_completed": "Orders Completed / Day",
    "acceptance_rate":  "Acceptance Rate",
}
```

```python
#get some preliminary data sorted 
trend_df    = driver_summary_grouped.sort_values("date")
all_regions = sorted(trend_df["region"].unique())
colormap      = cm.get_cmap("tab20b", len(all_regions))
REGION_COLORS = {region: colormap(i) for i, region in enumerate(all_regions)}
INCENTIVES = {'Active Bonus', 'Daily Productivity', 'Guarantee', 'No Incentive', 'Streak'}
```

```python
#get city grouping mapping 
incentive_region_df = pd.read_csv('../incentive_region_mapping.csv')
incentive_region_df
```

```python
#mapping incentives regions city grouping to a diction 
incentive_to_regions = {
    incentive: incentive_region_df[incentive_region_df["Incentive"] == incentive]["region"].tolist()
    for incentive in INCENTIVES
}
```

```python
for col, ylabel in metrics.items():
    fig, axes = plt.subplots(
        nrows=len(TIER_ORDER),
        ncols=len(INCENTIVES),
        figsize=(6 * len(INCENTIVES), 5 * len(TIER_ORDER)),
        sharex=True,
        sharey=True
    )

    for t_idx, tier in enumerate(TIER_ORDER):
        for i_idx, incentive in enumerate(INCENTIVES):
            ax = axes[t_idx, i_idx]

            # Get the regions that are the same as Week 1 (grouped by incentive) 
            regions_in_group = incentive_to_regions[incentive]  

            df_sub = trend_df[  
                (trend_df["Rider Group"] == tier) &
                (trend_df["Incentive"]   == "No Incentive") &  # all No Incentive now
                (trend_df["region"].isin(regions_in_group))    # filter to this group's regions
            ]

            for region in regions_in_group:
                df_r = df_sub[df_sub["region"] == region].sort_values("date")
                if df_r.empty:
                    continue
                ax.plot(
                    df_r["date"], df_r[col],
                    marker="o", markersize=3, linewidth=1.8,
                    color=REGION_COLORS[region],
                    label=region
                )

            if i_idx == 0:
                ax.set_ylabel(f"{tier}\n{ylabel}", fontsize=15)
            if t_idx == 0:
                ax.set_title(f"{incentive}\n(now No Incentive)", fontsize=15, fontweight="bold")

            ax.tick_params(axis="x", rotation=30, labelsize=15)
            ax.spines[["top", "right"]].set_visible(False)

    plt.tight_layout()

    for i_idx, incentive in enumerate(INCENTIVES):
        regions_in_col = incentive_to_regions[incentive]
        handles = [
            plt.Line2D([0], [0], color=REGION_COLORS[r], linewidth=1.5,
                       marker="o", markersize=4, label=r)
            for r in regions_in_col
        ]

        col_width = 1 / len(INCENTIVES)
        x_center = col_width * i_idx + col_width / 2

        fig.legend(
            handles=handles,
            title=incentive,
            loc="upper center",
            bbox_to_anchor=(x_center, -0.01),
            ncol=2,
            fontsize=9,
            title_fontsize=10,
            frameon=True,
            borderpad=0.6,
            handlelength=1.2,
            handletextpad=0.4,
            columnspacing=0.8
        )

    fig.suptitle(f"Week 0 — {ylabel} by City, Incentive Group & Driver Tier",
                 fontsize=20, fontweight="bold", y=1.01)

    plt.savefig(f"./week0_stats/week0_{col}.png", dpi=150, bbox_inches="tight")
    plt.show()
```

```python
# ── Config ────────────────────────────────────────────────────────────────────
sns.set_theme(style="whitegrid", font_scale=1.3)

TIER_ORDER = ["Top", "Mid", "Bottom"]
INCENTIVES = ["Daily Orders", "No Incentive", "Streak Orders", "Daily Hours", "Streak Hours"]
STATUS     = ["Win", "Lose", "Swing"]

metrics = {
    "hours_worked":     "Hours Worked / Day",
    "orders_completed": "Orders Completed / Day",
    "acceptance_rate":  "Acceptance Rate",
}

# ── Color by incentive type ────────────────────────────────────────────────────
INCENTIVE_COLORS = {
    "Daily Orders":           "#1f77b4",
    "No Incentive":        "#ff7f0e",
    "Streak Orders":        "#2ca02c",
    "Daily Hours":              "#d62728",
    "Streak Hours":  "#9467bd",
}

# ── Line style cycles by city name ────────────────────────────────────────────
LINE_STYLES = ["-", "--", "-.", ":", (0,(3,1,1,1)), (0,(5,2))]

trend_df    = driver_summary_grouped.sort_values("date")

# Build a stable linestyle mapping across ALL cities so the same city
# always gets the same dash pattern within a figure.
all_cities   = sorted(trend_df["region"].unique())
CITY_LSTYLES = {city: LINE_STYLES[i % len(LINE_STYLES)] for i, city in enumerate(all_cities)}

# You need a mapping: status → list of regions
# e.g. status_to_regions = {"Win": [...], "Lose": [...], "Swing": [...]}
# Adjust the column name "status" to whatever yours is called.
status_to_regions = (
    trend_df.drop_duplicates("region")
    .groupby("Status")["region"]
    .apply(list)
    .to_dict()
)

# ── Plotting ───────────────────────────────────────────────────────────────────
for status in STATUS:
    regions_in_status = status_to_regions.get(status, [])
    if not regions_in_status:
        continue

    for col, ylabel in metrics.items():
        fig, axes = plt.subplots(
            nrows=len(TIER_ORDER),
            ncols=len(INCENTIVES),
            figsize=(6 * len(INCENTIVES), 5 * len(TIER_ORDER)),
            sharex=True,
            sharey=True,
        )

        for t_idx, tier in enumerate(TIER_ORDER):
            for i_idx, incentive in enumerate(INCENTIVES):
                ax = axes[t_idx, i_idx]

                df_sub = trend_df[
                    (trend_df["Rider Group"] == tier) &
                    (trend_df["Incentive"]   == incentive) &
                    (trend_df["region"].isin(regions_in_status))
                ]

                for city in regions_in_status:
                    df_city = df_sub[df_sub["region"] == city].sort_values("date")
                    if df_city.empty:
                        continue
                    ax.plot(
                        df_city["date"],
                        df_city[col],
                        color=INCENTIVE_COLORS[incentive],
                        linestyle=CITY_LSTYLES[city],
                        marker="o",
                        markersize=3,
                        linewidth=1.8,
                        label=city,
                    )

                # Axis labels & titles
                if i_idx == 0:
                    ax.set_ylabel(f"{tier}\n{ylabel}", fontsize=13)
                if t_idx == 0:
                    ax.set_title(incentive, fontsize=13, fontweight="bold")

                ax.tick_params(axis="x", rotation=30, labelsize=11)
                ax.spines[["top", "right"]].set_visible(False)

        plt.tight_layout()

        # ── Per-column legends (city → line style) ────────────────────────────
        # One shared color legend (incentive) + one linestyle legend (city)

        # Incentive color legend  (bottom left)
        incentive_handles = [
            plt.Line2D([0], [0], color=INCENTIVE_COLORS[inc], linewidth=2, label=inc)
            for inc in INCENTIVES
        ]
        fig.legend(
            handles=incentive_handles,
            title="Incentive Type",
            loc="lower left",
            bbox_to_anchor=(0.0, -0.06),
            ncol=len(INCENTIVES),
            fontsize=9,
            title_fontsize=10,
            frameon=True,
        )

        # City linestyle legend  (bottom right)
        city_handles = [
            plt.Line2D(
                [0], [0],
                color="black",
                linestyle=CITY_LSTYLES[city],
                linewidth=2,
                label=city,
            )
            for city in sorted(regions_in_status)
        ]
        fig.legend(
            handles=city_handles,
            title="City",
            loc="lower right",
            bbox_to_anchor=(1.0, -0.06),
            ncol=3,
            fontsize=9,
            title_fontsize=10,
            frameon=True,
        )

        fig.suptitle(
            f"{status} Cities — {ylabel} by Incentive & Driver Tier",
            fontsize=18,
            fontweight="bold",
            y=1.02,
        )

        safe_status = status.lower()
        plt.savefig(
            f"./week_stats/{safe_status}_{col}.png",
            dpi=150,
            bbox_inches="tight",
        )
        plt.show()
        plt.close(fig)
```
