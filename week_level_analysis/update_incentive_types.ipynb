```python
import pandas as pd
import os
```

```python
folder_path_1 = '../progress_metrics/results_updated'
folder_path_2 = '../progress_metrics/daily_guarantee/daily_guarantee_results_updated'

output_path_1 = '../progress_metrics/results_updated_2'
output_path_2 = '../progress_metrics/daily_guarantee/daily_guarantee_results_updated_2'
```

```python
incentive_map = {
    'Guarantee': 'Daily Hours',
    'Active Bonus': 'Streak Hours',
    'Daily Productivity': 'Daily Orders',
    'Streak': 'Streak Orders',
}
```

```python
#add to all incentives datasets
for filename in os.listdir(folder_path_1): 
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join(folder_path_1, filename)) 
        df['Incentive'] = df['Incentive'].replace(incentive_map)
        df.to_csv(os.path.join(output_path_1, filename), index=False)
```

```python
#add to all incentives datasets
for filename in os.listdir(folder_path_2): 
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join(folder_path_2, filename))
        if 'Incentive' in df.columns: #08-06-08/15 no incentive data
            df['Incentive'] = df['Incentive'].replace(incentive_map)
        df.to_csv(os.path.join(output_path_2, filename), index=False)
```
