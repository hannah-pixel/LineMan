```python
incentives_mapping = {'guarantee_' : 'Daily Hours',
                      'active_bonus_' : 'Streak Hours', 
                      'streak_' : 'Streak Orders', 
                      'daily_': 'Daily Orders', 
                      'GUARANTEE': 'Daily Hours',
                      'ACTIVE_BONUS': 'Streak Hours', 
                      'PRODUCTIVITY': 'Daily Orders', #questionable...
                      }
```

```python
def get_incentive(row):
    for col_key, incentive_val in incentives_mapping.items():
        matching_cols = [col for col in row.index if col_key in col]
        for col in matching_cols:
            if pd.notna(row[col]) and row[col] != '' and row[col] != 0:
                return incentive_val
    return None
```

```python
folder_path = '../bkk'
output_folder = '../bkk_with_incentives'
```

```python
os.makedirs(output_folder, exist_ok=True)
```

```python
for filepath in glob.glob(os.path.join(folder_path, '*.csv')):  
    df = pd.read_csv(filepath) 

    df['incentive'] = df.apply(get_incentive, axis=1)

    file_name = os.path.basename(filepath)
    df.to_csv(os.path.join(output_folder, file_name), index=False)  
    print(f"Processed: {filepath}")
```
