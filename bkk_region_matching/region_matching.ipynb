```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.optimize import linear_sum_assignment
from sklearn.metrics import euclidean_distances
from scipy.spatial.distance import pdist, squareform
```

```python
#now get 1-1 mapping between bkk regions and 30 cities 

def get_region_profiles(df):
    # pivot so metrics become columns (Daily Wage_mean, Daily orders_mean, etc.)
    perf_metrics = df.pivot_table(
        index='region', 
        columns='metric', 
        values='mean'
    )

    #take number driver tier columns into consideration
    tier_cols = [c for c in df.columns if c.startswith('num_') or c == 'total_regional_drivers']
    tier_info = df.groupby('region')[tier_cols].first() #since all regions have the same counts

    #take mean for performation metrics into consideration (rename columns so don't clash with other data) 
    perf_metrics.columns = [f"mean_{col}" for col in perf_metrics.columns]
    
    combined_profile = pd.concat([perf_metrics, tier_info], axis=1)
    
    return combined_profile.fillna(0)
```

```python
def get_region_profiles(df):
    perf_metrics = df.pivot_table(
        index='region',
        columns='metric',
        values='mean'
    )
    perf_metrics.columns = [f"mean_{col}" for col in perf_metrics.columns]

    # isolate tier columns, collapse duplicates by summing across same-named columns
    tier_cols = [c for c in df.columns if c.startswith('num_') or c == 'total_regional_drivers']
    tier_df = df[['region'] + tier_cols].copy()
    tier_df = tier_df.T.groupby(level=0).sum().T  # sum duplicate columns
    tier_info = tier_df.groupby('region').first()

    combined_profile = pd.concat([perf_metrics, tier_info], axis=1)
    return combined_profile.fillna(0)
```

```python
# Prepare both profiles
profile_a = get_region_profiles(final_table_30_cities)
profile_b = get_region_profiles(final_table_bkk)

# align columns (ensure both DFs have the same features in the same order)
common_cols = profile_a.columns.intersection(profile_b.columns)
profile_a = profile_a[common_cols]
profile_b = profile_b[common_cols]

# standardize the data 
scaler = StandardScaler()
# Fit on a combination of both to ensure the same scale (since different numbers for differnt metrics, e.g. wage vs # orders could be very diff)
combined_data = pd.concat([profile_a, profile_b])
scaler.fit(combined_data)

scaled_a = scaler.transform(profile_a)
scaled_b = scaler.transform(profile_b)

# calculate Distance Matrix
# Rows = Regions in A, Columns = Regions in B w/ each cell a distance score 
distances = euclidean_distances(scaled_a, scaled_b)

# the "Hungarian Algorithm" for optimal 1-1 matching
# This finds the configuration that minimizes the total global distance
#so first city that matches its closest doesn't "steal" best match for second city
row_ind, col_ind = linear_sum_assignment(distances)

# extract the mapping
mapping_results = []
for i, j in zip(row_ind, col_ind):
    mapping_results.append({
        'Source_Region': profile_a.index[i],
        'Matched_Region': profile_b.index[j],
        'Similarity_Distance': distances[i, j]
    })

mapping_df = pd.DataFrame(mapping_results)
```

```python
#now try to map multiple of the different regions uniquely with one of the 30 cities (since 155 unique districts, and 30 unique cities)
```

```python
# Prepare both profiles
profile_cities = get_region_profiles(final_table_30_cities)
profile_districts = get_region_profiles(final_table_bkk)
```

```python
# Scale everything together
# Since columns are identical, we just stack them to find the global min/max/mean for scaling
common_cols = profile_districts.columns # They are the same, so no intersection needed
scaler = StandardScaler()
scaler.fit(pd.concat([profile_districts, profile_cities]))

scaled_districts = scaler.transform(profile_districts)
scaled_cities = scaler.transform(profile_cities)

# 2. Distance Matrix calculation
dist_matrix = euclidean_distances(scaled_districts, scaled_cities)
dist_df = pd.DataFrame(dist_matrix, index=profile_districts.index, columns=profile_cities.index)

# This turns the matrix into a long list of all possible pairs
stacked = dist_df.stack()
stacked.index.names = ['bkk_region_idx', 'city_30_idx'] # Unique names for the index levels
all_pairs = stacked.reset_index()
all_pairs.columns = ['bkk_region', 'city_30', 'distance']

# ort by most similar (lowest distance)
all_pairs = all_pairs.sort_values('distance')

# Since we have ~155 BKK regions (~93 with data) and 30 cities, set capacity to 4
city_capacities = {city: 4 for city in profile_cities.index}
assigned_bkk_regions = set()
mapping_results = []

for _, row in all_pairs.iterrows():
    bkk = row['bkk_region']
    city = row['city_30']
    
    # If this BKK region hasn't been matched yet AND the city has an open slot
    if bkk not in assigned_bkk_regions and city_capacities[city] > 0:
        mapping_results.append({
            'Bangkok_District': bkk,
            'Matched_30_City': city,
            'Similarity_Distance': row['distance']
        })
        assigned_bkk_regions.add(bkk)
        city_capacities[city] -= 1
    
    # Exit early if all 155 BKK regions are paired
    if len(assigned_bkk_regions) == len(profile_districts):
        break

mapping_df_multiple = pd.DataFrame(mapping_results)
```
