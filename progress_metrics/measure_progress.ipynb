```python
streak_results = streak_results.drop(['hour', 'time_worked_sec', 'Unnamed: 0.1', 'Unnamed: 0', 'is_accepted_target1', 'is_accepted_target2', 'is_accepted_target3', 'Start Date_x', 'End Date_x', 'Start Date_y', 'End Date_y'], axis=1)
```

```python
def count_matching_days(row):
    # Convert string to list by splitting on comma
    if pd.isna(row['days_worked_list']) or pd.isna(row['Active Hour Require Dow']):
        return 0
    
    # Split and clean the strings
    days_worked = [day.strip() for day in str(row['days_worked_list']).split(',')]
    required_dow = [day.strip() for day in str(row['Active Hour Require Dow']).split(',')]
    
    # Count matches
    return sum(1 for day in days_worked if day in required_dow)

streak_results['days_required_worked_count'] = streak_results.apply(count_matching_days, axis=1)
```

```python
streak_results.to_csv('./streak/streak_results/streak_results_with_incentives_cleaned.csv')
```

```python
conditions = [
    streak_results['Incentive'] == 'Daily Productivity',
    streak_results['Incentive'] == 'Active Bonus',
    streak_results['Incentive'] == 'Guarantee',
    streak_results['Incentive'] == 'Streak'
]

# We take average of all metrics to get progress towards completion of incentive
choices = [
    # Daily productivity 
    streak_results['daily_order_count'] / streak_results['Daily Productivity target'],
    
    # Active hour
    (
        (streak_results['daily_cumulative_time_hr'] / streak_results['Active Hour Required Online Hour']) +
        (streak_results['window_cumulative_hr'] / streak_results['Active Hour Required Hunting Hour']) +
        (streak_results['days_required_worked_count'] / 4) +
        (streak_results['days_worked_count'] / streak_results['Active Hour Required day'])
    ) / 4,
    
    # Daily guarantee
    (
        (streak_results['daily_cumulative_time_hr'] / streak_results['Daily Guarantee Required Online Hour']) +
        (streak_results['window_cumulative_hr'] / streak_results['Daily Guarantee Required Hunting Hour'])
    ) / 2,
    
    # Streak productivity
    (
        (streak_results['orders_target1_cumulative'] / streak_results['streak target P1']) +
        (streak_results['orders_target2_cumulative'] / streak_results['streak target P2']) +
        (streak_results['orders_target3_cumulative'] / streak_results['streak target P3'])
    ) / 3
]

streak_results['progress'] = np.select(conditions, choices, default=np.nan)
```
