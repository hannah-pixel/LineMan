```python
# Load streak_results
streak_file = Path('./streak/streak_results/streak_results.csv')
df_streak = pd.read_csv(streak_file)

# Convert date columns in streak_results
df_streak['date'] = pd.to_datetime(df_streak['date'])

# Load all 4 incentives files
incentives_dir = Path('../incentives')
incentive_files = [
    './incentives_7_14.csv',
    './incentives_7_21.csv', 
    './incentives_8_16.csv',
    './incentives_8_23.csv'
]

# Read and combine all incentive files
df_incentives_list = []
for file in incentive_files:
    file_path = incentives_dir / file
    df_temp = pd.read_csv(file_path)
    df_incentives_list.append(df_temp)

# Concatenate all incentive files
df_incentives = pd.concat(df_incentives_list, ignore_index=True)
```

```python
def parse_date_with_year(date_str, default_year=2025):
    if pd.isna(date_str):
        return pd.NaT
    
    try:
        # Parse the date
        date_str = str(date_str).strip()
        parsed = pd.to_datetime(date_str, format='%d-%b')
        
        # Replace year with default year if not year
        return parsed.replace(year=default_year)
    except:
        try:
            return pd.to_datetime(date_str, dayfirst=True, errors='coerce')
        except:
            return pd.NaT

default_year = 2025
# Parse dates
df_incentives['Start Date'] = df_incentives['Start Date'].apply(
    lambda x: parse_date_with_year(x, default_year)
)
df_incentives['End Date'] = df_incentives['End Date'].apply(
    lambda x: parse_date_with_year(x, default_year)
)
```

```python
merge_columns = [
    'Region',
    'Tier',
    'Start Date',
    'End Date',
    'Daily Guarantee Required Online Hour',
    'Daily Guarantee Hunting hour period',
    'Daily Guarantee Required Hunting Hour',
    'Daily Guarantee amount',
    'Active Hour Required Online Hour',
    'Active Hour Hunting hour period',
    'Active Hour Required Hunting Hour',
    'Active Hour Required day',
    'Active Hour Require Dow',
    'Active Hour Reward',
    'Daily Productivity target',
    'Daily Productivity Reward',
    'Streak target',
    'Streak Reward',
    'streak target P1',
    'streak reward P1',
    'streak target P2',
    'streak reward P2',
    'streak target P3',
    'streak reward P3'
]

# Keep only the columns we need from incentives 
available_merge_columns = [col for col in merge_columns if col in df_incentives.columns]
missing_columns = [col for col in merge_columns if col not in df_incentives.columns]

if missing_columns:
    print(f"\nWarning: These columns are not in incentives data: {missing_columns}")

df_incentives_subset = df_incentives[available_merge_columns].copy()
```

```python
merged_df = df_streak.merge(
    df_incentives_subset,
    left_on=['region', 'Tier'],
    right_on=['region', 'Tier'],
    how='left'  # Keep all rows from streak_results
)

merged_df['Start Date'] = merged_df['Start Date_y']
merged_df['End Date'] = merged_df['End Date_y']
merged_df.drop

# Create a mask for rows within the date range OR rows with no incentive match
date_range_mask = (
    (merged_df['date'] >= merged_df['Start Date_y']) & 
    (merged_df['date'] <= merged_df['End Date_y'])
) | (merged_df['Start Date_y'].isna())  # Keep rows with no incentive match

merged_df = merged_df[date_range_mask].copy()
```
