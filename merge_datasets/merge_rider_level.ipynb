```python
from pathlib import Path
import os
```

```python
# get all CSV files
csv_files = sorted(Path('../mainexperiment/weekly/weather_merged').glob('*.csv'))

# display all files
for i, file in enumerate(csv_files, 1):
    print(f"{i}. {file.name}")

# prepare drivers dataframe 
drivers['Start Date'] = pd.to_datetime(drivers['Start Date']).dt.date
drivers['End Date'] = pd.to_datetime(drivers['End Date']).dt.date

# process each file
for i, csv_file in enumerate(csv_files, 1):
    print(f"\n[{i}/{len(csv_files)}] Processing: {csv_file.name}")
    print("-" * 80)
    
    # read the CSV file
    df = pd.read_csv(csv_file)
    df['date'] = pd.to_datetime(df['date']).dt.date
    
    # merge based on date range condition
    df_merged = df.merge(
        drivers,
        on=['region', 'driver_id'],
        how='left'
    )
    
    # Filter to keep only rows where date falls within the range
    df_merged = df_merged[
        (df_merged['date'] >= df_merged['Start Date']) & 
        (df_merged['date'] <= df_merged['End Date'])
    ].copy()
        
    print(f"Rows after filtering by date range: {len(df_merged)}")
    
    output_name = f"tiers_{csv_file.name}"
    output_path = os.path.join('../mainexperiment/weekly/tier_merged', output_name)
    
    # Save to CSV
    df_merged.to_csv(output_path, index=False)
```

```python

DATA_DIR        = "../mainexperiment/weekly/tier_merged"             # folder containing your CSV files
W3_PATTERN      = "*0721-0727*"   # filename pattern for W3
W4_PATTERN      = "*0728-0810*"   # filename pattern for W4
NO_INCENTIVE    = "No Incentive"  # value to set in Incentive column

DRIVER_COL      = "driver_id"
TIER_COL        = "driver_tier"
INCENTIVE_COL   = "Incentive"
```

```python
"""
Fix W4 (0728_0810):
  1. Overwrite driver_tier with each driver's tier from W3 (0721_0727)
  2. Set all Incentive values to 'No Incentive'
  3. Overwrite the original W4 file
"""

import os
import glob
import pandas as pd

# ═══════════════════════════════════════════════════
# CONFIG
# ═══════════════════════════════════════════════════

DATA_DIR        = "../mainexperiment/weekly/tier_merged"             # folder containing your CSV files
W3_PATTERN      = "*0721-0727*"   # filename pattern for W3
W4_PATTERN      = "*0728-0810*"   # filename pattern for W4
NO_INCENTIVE    = "No Incentive"  # value to set in Incentive column

DRIVER_COL      = "driver_id"
TIER_COL        = "Rider Group"
INCENTIVE_COL   = "Incentive"

# ═══════════════════════════════════════════════════
# LOAD
# ═══════════════════════════════════════════════════

def find_file(data_dir, pattern):
    matches = glob.glob(os.path.join(data_dir, pattern))
    if not matches:
        raise FileNotFoundError(f"No file found matching '{pattern}' in {data_dir}")
    if len(matches) > 1:
        raise ValueError(f"Multiple files matched '{pattern}': {matches}")
    return matches[0]

w3_path = find_file(DATA_DIR, W3_PATTERN)
w4_path = find_file(DATA_DIR, W4_PATTERN)

print(f"W3 file: {w3_path}")
print(f"W4 file: {w4_path}")

w3 = pd.read_csv(w3_path, low_memory=False)
w4 = pd.read_csv(w4_path, low_memory=False)

print(f"\nW3 rows: {len(w3):,}")
print(f"W4 rows: {len(w4):,}")

# ═══════════════════════════════════════════════════
# STEP 1: Build driver → tier lookup from W3
# ═══════════════════════════════════════════════════

tier_lookup = (
    w3.dropna(subset=[TIER_COL])
    .groupby(DRIVER_COL)[TIER_COL]
    .agg(lambda x: x.mode().iloc[0])
    .to_dict()
)

print(f"\nDrivers in W3 tier lookup: {len(tier_lookup):,}")

# ═══════════════════════════════════════════════════
# STEP 2: Overwrite driver_tier in W4
# ═══════════════════════════════════════════════════

new_tiers = w4[DRIVER_COL].map(tier_lookup)
n_matched  = new_tiers.notna().sum()
n_missing  = new_tiers.isna().sum()

w4[TIER_COL] = new_tiers  # create column from scratch using W3 lookup
w4['Start Date'] = '2025-07-28'
w4['End Date'] = '2025-08-10'

no_match = w4['Start Date'].isna()
in_range = (
        (w4['date'] >= pd.to_datetime(w4['Start Date'])) & 
        (w4['date'] <= pd.to_datetime(w4['End Date']))
    )
    
w4 = w4[in_range | no_match].copy() 


print(f"driver_tier mapped from W3: {n_matched:,}")
print(f"driver_tier not found in W3 (will be NaN): {n_missing:,}")

# ═══════════════════════════════════════════════════
# STEP 3: Set all Incentive = 'No Incentive'
# ═══════════════════════════════════════════════════

w4.loc[:, INCENTIVE_COL] = NO_INCENTIVE
print(f"\nIncentive set to '{NO_INCENTIVE}' for all {len(w4):,} rows")

# ═══════════════════════════════════════════════════
# SAVE — overwrite original W4 file
# ═══════════════════════════════════════════════════

w4.to_csv(w4_path, index=False)
print(f"\n✓ Saved (overwritten): {w4_path}")
```
