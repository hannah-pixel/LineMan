```python
import pandas as pd
```

```python
weather = pd.read_csv('./mainexperiment/rain-0710-0829.csv')
```

```python
weather['region'] = weather['region::multi-filter']
```

```python
from pathlib import Path
import os
```

```python
# get all CSV files
csv_files = sorted(Path('./mainexperiment/weekly').glob('*.csv'))

# display all files
for i, file in enumerate(csv_files, 1):
    print(f"{i}. {file.name}")
    
# Process each file
for i, csv_file in enumerate(csv_files, 1):
    print(f"\n[{i}/{len(csv_files)}] Processing: {csv_file.name}")
    print("-" * 80)
    
    df = pd.read_csv(csv_file)
    df_merged = pd.merge(df, weather,
                on=['region', 'date'],
                how='left')

    output_name = f"merged_{csv_file.name}"
    output_path = os.path.join('./mainexperiment/weekly/weather_merged', output_name)
    
    # save to CSV
    df_merged.to_csv(output_path, index=False)
```
